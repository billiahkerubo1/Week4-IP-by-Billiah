```python
#importing our libraries
import pandas as pd
import numpy as np
```

```python
#reading the dataset and creating dataframe
df = pd.read_csv('/content/Autolib_dataset (2).csv', parse_dates = ["Scheduled at"] , delimiter=',')
df
```

```python
#previewing the dataset using the first 5 rows
df.head()
```

```python
#accessing the dataset information
df.info()
```

```python
#describing the shape of the dataframe
df.shape
```

```python
#for uniformity, convert columns to uppercase
df.columns = map(str.upper, df.columns)
df
```

```python
#checking for any missing values
df.isnull().any()
```

```python
#Cleaning the dataset before analysis
#procedure 1: checking for null values
df.isnull().sum()
```

```python
#adjusting the columns with missing data
new_df = df.fillna({
    'Displayed comment' : 'no comment',
    'Scheduled at' : 'no schedule'
   });
(new_df)
```

```python
#checking if the missing data has been corrected

new_df.isnull().any()
```

```python
#the basic statistical information
new_df.describe()
```

```python
#Paris city records in April and over in 2018
d_paris = new_df[(new_df['CITY'] == 'Paris') & (new_df['MONTH'] >= 4) & (new_df['YEAR']==2018)]
d_paris
```

```python
#getting the hours of the Bluecar counter in Paris city
h_d_paris = d_paris[['ID','BLUECAR COUNTER','HOUR']]
h_d_paris
```

```python
#getting the most popular hour of Bluecars in Paris
h_d_paris.sort_values(by = 'HOUR', ascending = False)
```

```python
#overall most popular station
new_df.sort_values(by = 'STATION TYPE', ascending = False)
```

```python
#finding the most popular hour
pop_hr = new_df.sort_values( by = 'HOUR', ascending = False)
pop_hr
```

```python
#finding popular station on popular hour
pop_st = pop_hr.sort_values(by = 'STATION TYPE')
pop_st
```

```python
#getting the popular postal code for bluecars
b_pop_pc = new_df[['ID','BLUECAR COUNTER','POSTAL CODE', 'SLOTS']].sort_values(by = 'SLOTS', ascending = False)
b_pop_pc
```

```python
#checking if the popular station belongs in popular postal code
pop_st['STATION TYPE'].isin(['POSTAL CODE'])

```

```python
#checking for popular hour using utility counter in Paris city in April
utilib_pop_hr = d_paris[['ID', 'UTILIB COUNTER','HOUR']].sort_values (by = 'HOUR', ascending = False)
utilib_pop_hr
```

```python
#checking for most popular hour for Utilib 1.4 cars in Paris in April
utilib2_cars = d_paris[['ID', 'UTILIB 1.4 COUNTER','HOUR']].sort_values (by = 'HOUR', ascending = False)
utilib2_cars
```
